# 토픽과 밸류로 테이블 만들기

# 카프카에서 생성한 데이터를 구독해서 일정한 시간 단위로 사용자가 정의한 테이블에 저장하기

```python
from pyspark.sql import SparkSession, streaming
from pyspark.sql.functions import col,expr

# create session
spark = SparkSession.builder.appName('kafka_spark').getOrCreate()
```

```python
# DB 저장하기
def foreach_batch_function(batch_df,batch_id): 
    url = "jdbc:mysql://localhost:3306/hadoopguide?useSSL=false"
    driver='com.mysql.jdbc.Driver'
    user='root'
    password='password'
    tablename = 'tbl_kafka_spark'
    # mode = 'overwrite'
    mode = 'append'
    props = {'driver':driver,'user':user,'password':password}
    
    params = {
        'url': url,
        'table' :  tablename,    
        'properties' : props,
        'mode' : mode
    }    
    batch_df.write.jdbc(**params)   
```

```python
kafka_options = {
    'kafka.bootstrap.servers':'localhost:9092',
    'subscribe':'topic1,topic2'
}
df = spark.readStream.format("kafka")\
                    .options(**kafka_options).load()
rawDF = df.selectExpr("topic","CAST(value as STRING)")
topic2DF = rawDF.filter(col("topic") == "topic2")

              
#토픽2만 나오게하기
query = topic2DF.writeStream \
        .format("console") \
        .trigger(processingTime='5 seconds') \
        .foreachBatch(foreach_batch_function) \
        .start()  
query.awaitTermination()
```

# 토픽 별 다른 테이블에 저장

```python
def foreach_batch_function(batch_df,batch_id): 
    url = "jdbc:mysql://localhost:3306/hadoopguide?useSSL=false"
    driver='com.mysql.jdbc.Driver'
    user='root'
    password='password'
    tablename = 'tbl_kafka_spark'
    # mode = 'overwrite'
    mode = 'append'
    props = {'driver':driver,'user':user,'password':password}
    
    params = {
        'url': url,
        'table' :  tablename,    
        'properties' : props,
        'mode' : mode
    }    
    topic1DF = batch_df.filter(col("topic") == "topic1")
    topic2DF = batch_df.filter(col("topic") == "topic2")
    
    params['table'] = 'tbl_kafka_spark_topic1'
    topic1DF.write.jdbc(**params)   
    params['table'] = 'tbl_kafka_spark_topic2'
    topic2DF.write.jdbc(**params)   
```

```python
kafka_options = {
    'kafka.bootstrap.servers':'localhost:9092',
    'subscribe':'topic1,topic2'
}
df = spark.readStream.format("kafka")\
                    .options(**kafka_options).load()
rawDF = df.selectExpr("topic","CAST(value as STRING)")
              
#토픽별로 테이블 다르게 나오게하기
query = rawDF.writeStream \
        .format("console") \
        .trigger(processingTime='5 seconds') \
        .foreachBatch(foreach_batch_function) \
        .start()  
query.awaitTermination()
```

# GPT코드

```python
from pyspark.sql.functions import col

# 트리거 이용한 처리
def parseData(spark, df):
    # DataFrame의 'value' 컬럼을 선택하고, 이를 문자열로 변환
    result_df = df.selectExpr("CAST(value AS STRING) as value")
    return result_df
                                        

def proccess_batch(batchDF, batch_id):
    # 임시로 저장된 DataFrame을 메모리에서 해제
    batchDF.persist()

    # 'topic1' 토픽에 해당하는 데이터만 필터링
    topic1DF = batchDF.filter(col("topic") == "topic1")

    # 데이터 파싱
    parsedDF = parseData(spark, topic1DF)

    # 결과 출력
    parsedDF.show()

    # 임시로 저장된 DataFrame을 메모리에서 해제
    batchDF.unpersist()

# Kafka 옵션 설정 (예시입니다, 실제 옵션에 맞게 조정하세요)
kafka_options = {
    "kafka.bootstrap.servers": "localhost:9092",
    "subscribe": "topic1,topic2"
}

# Kafka에서 스트리밍 데이터 읽기
df = spark.readStream.format("kafka").options(**kafka_options).load()

# 'topic', 'key', 'value' 선택 및 데이터 타입 변환
rawDF = df.selectExpr("topic", "CAST(value AS STRING) as value")

# 스트리밍 쿼리 실행
query = rawDF.writeStream.trigger(processingTime='5 seconds').foreachBatch(proccess_batch).start()

# 스트리밍 쿼리 종료 대기
query.awaitTermination()
```
