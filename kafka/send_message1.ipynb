```python
from kafka import KafkaProducer, KafkaConsumer
import numpy as np
import os
import json
from time import time, sleep
```

```python
temp = [np.random.normal() for i in range(100)]
```

```python
producer = KafkaProducer(bootstrap_servers='localhost:9092', value_serializer=lambda x: json.dumps(x).encode("utf-8"))
```

```python
for data in temp:
    producer.send("random_data",value= data)
producer.flush()
```

```python
consumer = KafkaConsumer("random_data", bootstrap_servers='localhost:9092', value_deserializer= lambda x: json.loads(x.decode("utf-8"))
                         ,auto_offset_reset='earliest', consumer_timeout_ms=1000*5, group_id = 'g1')
```

```python
for data in consumer:
    print(data.value)
consumer.commit()
```

```python
for idx, data in enumerate(consumer):
    print(idx)
consumer.commit()
```
