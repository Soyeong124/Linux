```python
# 컨슈머 생성(토픽 데이터 소비)
from kafka import KafkaConsumer
import json
import time
from csv import reader
import threading
```

```python
class MessageConsumer:
    def __init__(self, broker, topic, group_id):
        self.broker = broker
        self.topic = topic
        self.group_id = group_id
    def active_listener(self):
        params = {
            "bootstrap_servers":self.broker,
            "group_id": self.group_id,
            "consumer_timeout_ms": 1000 * 5, # 5초
            "auto_offset_reset":"earliest", 
            #auto_offset_reset 토픽을 처음 구독할 때 
            #컨슈머의 오프셋이 브로커에 의해 삭제될 때 시작 오프셋을 설정
            "enable_auto_commit":False,
            #역직렬화
            #ASCII로 디코딩 JSON 형태로 파싱
            "value_deserializer": lambda x: json.loads(x.decode("utf-8")) if x else None
        }               
        consumer = KafkaConsumer(**params)
        consumer.subscribe(self.topic) #구독은 지정한 토픽에 데이터가 감지되면 읽는다(소비)
        #데이터 출력 - 데이터를 읽었을 때 저리할 로직
        with open(f"new-{self.topic}-consumer.txt","a")as f:  #w로 하면 오류가 발생함 a 로 적어야함 'w' 모드로 파일을 열 경우, 파일이 이미 존재한다면 그 내용을 지우고 새로 쓰기 시작
            for data in consumer:
                print(f'Topic : {data.topic}, Partition : {data.partition}, Offset : {data.offset}, Key : {data.key}, value : {data.value}')
                # print(data.value)
                f.write(f'Topic : {data.topic}, Partition : {data.partition}, Offset : {data.offset}, Key : {data.key}, value : {data.value}')
```

```python
broker = 'localhost:9092'
topic = 'TOPIC1'
group_id = "test2"
consumer1 = MessageConsumer(broker, topic, group_id)
consumer2 = MessageConsumer(broker, topic, group_id)
consumer3 = MessageConsumer(broker, topic, group_id)

consumer1.active_listener()
consumer2.active_listener()
consumer3.active_listener()
```

```python
broker = 'localhost:9092'
# topic = 'TOPIC1'
group_id = "test2"
consumer1 = MessageConsumer(broker, 'TOPIC1-1', group_id)
consumer2 = MessageConsumer(broker, 'TOPIC1-2', group_id)
consumer3 = MessageConsumer(broker, 'TOPIC1-3', group_id)

# 각 컨슈머의 active_listener 메서드를 별도의 스레드에서 실행
t1 = threading.Thread(target=consumer1.active_listener)
t2 = threading.Thread(target=consumer2.active_listener)
t3 = threading.Thread(target=consumer3.active_listener)

t1.start()
t2.start()
t3.start()

t1.join()
t2.join()
t3.join()
```
